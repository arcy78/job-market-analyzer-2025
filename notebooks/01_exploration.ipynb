```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Chargement des donnes
df = pd.read_csv('../data/raw/jobs_data.csv')

# Aperu
print(f"Nombre d'offres : {df.shape[0]}")
display(df.head())

# 2.Analyse des titres de poste les plus frquents
plt.figure(figsize=(10, 6))
top_titles = df['title'].value_counts().head(10)
sns.barplot(x=top_titles.values, y=top_titles.index, palette="magma")

plt.title("Top 10 des types de postes (Fake Jobs)")
plt.xlabel("Nombre d'offres")
plt.show()

# 3. Recherche de mots-cls (Simulation)
# On regarde combien d'offres contiennent "Python" vs "HTML" dans le titre (juste pour l'exemple)
python_jobs = df[df['title'].str.contains("Python", case=False)].shape[0]
html_jobs = df[df['title'].str.contains("HTML", case=False)].shape[0]

print(f"Offres mentionnant Python : {python_jobs}")
print(f"Offres mentionnant HTML : {html_jobs}")
```
